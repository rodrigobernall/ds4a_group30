```python
uber1['YYYY'] = pd.DatetimeIndex(uber1['pickup_datetime']).year
uber1['M'] = pd.DatetimeIndex(uber1['pickup_datetime']).month
uber1['D'] = pd.DatetimeIndex(uber1['pickup_datetime']).day
uber1["YYYYMM"]=uber1.YYYY.astype(str) + uber1.M.astype(str)
uber1["YYYYMD"]=uber1.YYYY.astype(str) + uber1.M.astype(str)+ uber1.D.astype(str)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
uber1.groupby('YYYYMD')["pickup_datetime"].count().plot(ax=ax)
```

```python
uber1NP=uber1.groupby(["YYYYMD","YYYY","M","D"])["pickup_datetime"].count()
```

```python
uber1NP.to_csv("uber1NP.csv",  header=True)
```

```python
uber2['YYYY'] = pd.DatetimeIndex(uber2['pickup_datetime']).year
uber2['M'] = pd.DatetimeIndex(uber2['pickup_datetime']).month
uber2['D'] = pd.DatetimeIndex(uber2['pickup_datetime']).day
uber2["YYYYMM"]=uber2.YYYY.astype(str) + uber2.M.astype(str)
uber2["YYYYMD"]=uber2.YYYY.astype(str) + uber2.M.astype(str)+ uber2.D.astype(str)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
uber2.groupby('YYYYMD')["pickup_datetime"].count().plot(ax=ax)
```

```python
uber1NP=uber1.groupby(["YYYYMD","YYYY","M","D"])["pickup_datetime"].count()
uber2NP=uber2.groupby(["YYYYMD","YYYY","M","D"])["pickup_datetime"].count()
```

```python
uber1NP.to_csv("uber1NP.csv",  header=True)
uber2NP.to_csv("uber2NP.csv",  header=True)
```
